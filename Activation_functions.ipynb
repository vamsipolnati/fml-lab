**Activation Functions**

Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
x= np.linspace(-10,10)
```

```python
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

**Tan h**

```python
def tanh(x):
  return np.tanh(x)
```

```python
x= np.linspace(-10,10)
```

```python
plt.plot(x,tanh(x))
plt.axis('tight')
plt.title('Activation Function :Tanh')
plt.show()
```
